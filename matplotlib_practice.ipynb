```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Create a Line Chart Using the MATLAB Method

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create a Line Chart Using the Object-Oriented Interface
# Create the plot with ax.plt()

fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# or can be like this too
# Create the plot with ax.plt()

fig = plt.figure()
ax = fig.add_subplot()
```

```python
#Add this 
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a lable for the legend
plt.plot(x_axis, y_axis, label='Boston')

# Create lables for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45
plt.ylim(0, 45)

#Create a title
plt.title("PyBer Fare by Month")

#Add the legend
plt.legend()
```

```python
# Add new line color, modified line width, and marker for data points
# Create th eplot
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=3, label="Boston")
# Create lables for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45
plt.ylim(0, 45)
#ax.grid()
#Create a title
plt.title("PyBer Fare by Month")
plt.grid()
#Add the legend
plt.legend()
```

```python
# Using the object-oriented approach

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create Scatter Plots and Bubble Charts
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# When we use the plt.plot() function to create a scatter plot,
#we need to add a lowercase "o" as a parameter inside the parentheses. 
#This switches the plot from a line chart to a scatter plot.

plt.plot(x_axis, y_axis, 'o')
```
